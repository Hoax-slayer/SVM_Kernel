```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = df.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(df.values, i)
                          for i in range(len(df.columns))]
```

```python
print(vif_data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= df
y=df_cancer['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='linear')
```

```python
X_train
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf','linear']} 
```

```python
grid= GridSearchCV(SVC(),param_grid,verbose=4,refit=True)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
grid_predictions= grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
print('\n')
print(classification_report(y_test,grid_predictions))
```

```python
new_model=SVC(C=100,kernel='linear')
```

```python
new_model.fit(X_train,y_train)
```

```python
new_predictions = new_model.predict(X_test)
```

```python
print(confusion_matrix(y_test,new_predictions))
print('\n')
print(classification_report(y_test,new_predictions))
```
